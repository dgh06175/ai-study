```python
from tensorflow.keras.applications.xception import Xception

# 기본 입력 크기(299, 299, 3), 1000 클래스 분류기 포함
model_xception = Xception()
model_xception.summary()
```

```python
from tensorflow.keras import datasets

(train_x, train_y), (test_x, test_y) = datasets.fashion_mnist.load_data()
```

```python
import cv2
import numpy as np

def myResize(input_imgs, dst_size):
    out_imgs = []
    for img in input_imgs:
        img_out = cv2.resize(img, dsize=dst_size, interpolation=cv2.INTER_LINEAR)
        img_out = cv2.merge([img_out, img_out, img_out])  # 흑백 → 3채널
        out_imgs.append(img_out)
    return np.array(out_imgs)
```

```python
train_x_299 = myResize(train_x, 299)
test_x_299 = myResize(test_x, 299)
```

```python
train_x_299.shape, train_x.shape
```

```python
import tensorflow as tf

model_xception.compile(
    optimizer=tf.keras.optimizers.Adam(),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

num_epochs = 1
batch_size = 32
hist = model_xception.fit(train_x_299, train_y, batch_size=batch_size, shuffle=True, epochs=num_epochs)
```

```python
test_x_0_to_1000 = test_x[:1000]
test_x_299x_299 = myResize(test_x_0_to_1000, (299, 299))
test_x_299x_299 = test_x_299x_299 / 255.
test_x_0_to_1000 = test_y[:1000]
test_x_299.shape
```
