```python
class Counter:
    def __init__(self, num):
        self.num = num

    def increment(self):
        self.num += 1

    def reset(self):
        self.inputValue(0)

    def inputValue(self, num):
        self.num = num

    def printCurrentValue(self):
        print("value: ", self.num)

c1 = Counter(0)
c2 = Counter(10)
c1.printCurrentValue()
c2.printCurrentValue()

c1.increment()
c1.increment()
c2.increment()

c1.printCurrentValue()
c2.printCurrentValue()
    
```

```python
class Point2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __add__ (self, pt):
        new_x = self.x + pt.x
        new_y = self.y + pt.y
        return Point2D(new_x, new_y)
    def __sub__ (self, pt):
        new_x = self.x - pt.x
        new_y = self.y - pt.y
        return Point2D(new_x, new_y)
    def __str__(self):
        return f"({self.x}, {self.y})"
    
p1 = Point2D(3, 5)
p2 = Point2D(1, -1)

p3 = p1 + p2
p4 = p1 - p2

print(p1)
print(p2)
print()
print("add =", p3)
print("sub =", p4)
```

```python
import math

class ComplexNumber:
    def __init__(self, real, img):
        self.real = real
        self.img = img
    
    def __str__(self):
        if self.img >= 0:
            return f"{self.real}+j{self.img}"
        else:
            return f"{self.real}-j{abs(self.img)}"
        
    def __add__(self, cn):
        return ComplexNumber(self.real + cn.real, self.img + cn.img)
    
    def __sub__(self, cn):
        return ComplexNumber(self.real - cn.real, self.img - cn.img)
    
    def __mul__(self, x):
        if type(x) == int:
            return ComplexNumber(self.real * x, self.img * x)
        elif type(x) == type(self):
            return ComplexNumber(self.real * x.real - self.img * x.img , self.img * x.real + self.real * x.img)
        else:
            print(f"INVALID TYPE: {type(x)}")
    
    def __eq__(self, cn):
        return self.real == cn.real and self.img == cn.img
    
    def __ne__(self, cn):
        return self.real != cn.real or self.img != cn.img
    
    def __abs__(self):
        return math.sqrt(self.real ** 2 | self.img ** 2)

    
a = ComplexNumber(1, 2)
b = ComplexNumber(-3, 3)
c = a + b
d = a - b
e = a * b
f = (a == b)
g = (a != b)
h = abs(a)
print(a, b, c, d, e, f, g, h, sep=", ")
```
